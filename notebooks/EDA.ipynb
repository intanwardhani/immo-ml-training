## 3.1. House price exploratory visualisation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load processed train data
processed_path = "../data/processed/train.csv"
train_df = pd.read_csv(processed_path)

# Set seaborn style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
# Raw price
plt.figure()
sns.histplot(train_df['price'], bins=50, kde=True, color='skyblue')
plt.title("House Price Distribution")
plt.xlabel("Price (€)")
plt.ylabel("Number of properties")
plt.show()

# Log-transformed price
plt.figure()
sns.histplot(np.log1p(train_df['price']), bins=50, kde=True, color='orange')
plt.title("Log-Transformed House Price Distribution")
plt.xlabel("log(1 + Price)")
plt.ylabel("Number of properties")
plt.show()
```

## 3.7. Identify outliers

```python
# Simple outlier detection: values outside 1.5*IQR
def detect_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    return outliers

for col in ['living_area', 'price']:
    outliers = detect_outliers(train_df, col)
    print(f"{col}: {len(outliers)} outliers ({len(outliers)/len(train_df)*100:.2f}%)")
```

# 4. Basic insights

### 4.1. Correlation interpretation
Feature `number_bedrooms` has the highest correlation coefficient (0.43) with `price`. But I wouldn't say they are highly correlated.

### 4.2. Skewness interpretation
* **0**   → perfectly symmetric
* **> 0** → right-skewed (tail to the right, high values are rare)
* **< 0** → left-skewed (tail to the left, low values are rare)


| Feature             | Skewness | Interpretation                                                                               |
| ------------------- | -------- | -------------------------------------------------------------------------------------------- |
| **facades**         | 0.19     | Approximately symmetric, no action needed.                                                   |
| **garden**          | -0.13    | Approximately symmetric, fine as-is.                                                         |
| **living_area**     | 98.44    | Extremely right-skewed! Most properties have small living areas, a few have very large ones. |
| **number_bedrooms** | 3.68     | Moderately right-skewed, but not extreme. Many houses with few bedrooms, some with many.     |
| **swimming_pool**   | 5.99     | Highly right-skewed, most houses have no pool (0), a few have a pool (1).                    |
| **terrace**         | -1.67    | Moderate left-skew; many properties have terraces (1), fewer don’t (0).                      |
| **price**           | 6.07     | Strongly right-skewed: most houses have moderate prices, some are very expensive.            |

**Insight:**

* Skewness is extreme for `living_area`, `swimming_pool`, and `price`.
* For ML models like Ridge regression, strong skew can hurt performance because the model assumes roughly symmetric errors. Tree-based models (RF/XGBoost) are less sensitive.

### 4.3. Outliers interpretation
* `living_area`: 526 outliers (4.84%)
  * Usually extremely large houses compared to the bulk of the data.
  * Could dominate regression fits for linear models.
* `price`: 731 outliers (6.73%)
  * Very expensive properties relative to most of the market.

**Insight:**

* Outliers are not huge percentages, but they are significant for linear models.
* Tree-based models (RandomForest/XGBoost) are robust to these.
* For Ridge, I will consider **capping/extreme value transformation** or **log-transforming price**.

# 5. Next steps

1. **Target transformation:**

   * Apply `y = log(price)` to reduce right-skew for regression.
   * This stabilizes variance and reduces outlier impact.

2. **Feature transformation:**

   * `living_area` could also be log-transformed for Ridge linear model.
   * `number_bedrooms` and `facades` likely fine as-is.

3. **Outlier handling:**

   * I will not remove outliers blindly, especially for tree models.
   * For Ridge linear model: I will do capping at 1–99 percentile or maybe transformations.

4. **Visualisation:**

   * Boxplots and histograms before and after log/capping to confirm if skew/outliers are handled.



# 6. Test before and after log-transformation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew

# Check skewness
print("Skewness of numeric features:")
for col in numeric_cols:
    print(f"{col}: {skew(train_df[col].dropna()):.2f}")
```

```python
# Plot histogram & boxplot
def plot_distribution(data, column, title_prefix="Before"):
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    sns.histplot(data[column], kde=True, ax=axes[0])
    axes[0].set_title(f"{title_prefix} Histogram: {column}")

    sns.boxplot(x=data[column], ax=axes[1])
    axes[1].set_title(f"{title_prefix} Boxplot: {column}")

    plt.show()

# BEFORE transformation
print("\n=== BEFORE TRANSFORMATION ===")
plot_distribution(train_df, "price")
plot_distribution(train_df, "living_area")
```

```python
# Apply log-transform
train_df["price_log"] = np.log1p(train_df["price"])               # safe log (log(1+x))
train_df["living_area_log"] = np.log1p(train_df["living_area"])   # OPTIONAL

print("\nSkewness AFTER log-transform:")
print(f"price_log: {skew(train_df['price_log'].dropna()):.2f}")
print(f"living_area_log: {skew(train_df['living_area_log'].dropna()):.2f}")
```

```python
# Visualise AFTER transformation
print("\n=== AFTER TRANSFORMATION ===")
plot_distribution(train_df, "price_log", title_prefix="After")
plot_distribution(train_df, "living_area_log", title_prefix="After")
```

```python
# Compare outliers before/after
def count_outliers(column):
    Q1 = train_df[column].quantile(0.25)
    Q3 = train_df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    return ((train_df[column] < lower) | (train_df[column] > upper)).sum()
print("\nOutlier counts:")
print(f"price before: {count_outliers('price')}")
print(f"price after log-transform: {count_outliers('price_log')}")
print(f"living_area before: {count_outliers('living_area')}")
print(f"living_area after log-transform: {count_outliers('living_area_log')}")
```
